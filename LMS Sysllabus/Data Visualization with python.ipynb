# Box Plot Outliers

Understanding box plot it is very difficult just by looking at them.

We will also understand the term called outliers and how boxplot will detect those outliers.

We will also understand the boxes and the lines inside the boxes and how to interpret let's watch the video.

--------------------------------------------------------------------------
Why are box plots useful? Box plots divide the data into sections that each contain approximately 25% of the data in that set. Box plots are useful as they provide a visual summary of the data enabling researchers to quickly identify mean values, the dispersion of the data set, and signs of skewness.

```python
emp={'Age_years':[10,23,22,25,20,23,24,27,25,24,33,21],'Salary_thousands':[15,22,24,25,30,23,34,43,19,20,52,21]}
df=pd.DataFrame(data=emp)
df.head()
```

```python
df.plot(kind='box',figsize=(8,5),color='r',yticks=range(0,55,5))
#yticks is the requent range
```

```python
sal=[15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

n=12 iteam 25%=(25/100)*(n+1)=3.25

Outliers

```python
# 1st quantile
# assuming sal is a list of numbers
sal_series = pd.Series(sal)

# 1st quantile
q1 = sal_series.quantile(0.25)
print('Q1:', q1)
```

```python
#3rd quantile
q3 = sal_series.quantile(0.75)
print('Q3:', q3)
```

```python
# find the IQR Inter quartile range
# identify outliers
iqr=q3-q1
iqr # it will help find the outliers
```

Qutliers Detecton formula

highter side= Q3+(1.5*IQR)

lower side=Q1-(1.5*IQR)


```python
upside=q3+(1.5*iqr)
print(upside)
```

```python
# anything above 46.375 is outliers

lower_side=q1-(1.5*iqr)
print(lower_side)
```

# Scatter matrix

Suppose you want to compare one variable with an another variable relationship. It is very difficult to plot 1 versus 1 what if we can apply some technique which will help us to compare one variable versus all other variable relationship let's look into scatter Matrix plot.

scatter_matrix() can be used to easily generate a group of scatter plots between all pairs of numerical features. It creates a plot for each numerical feature against every other numerical feature and also a histogram for each of them.

```python
from pandas.plotting import scatter_matrix
```

```python
iris = sns.load_dataset('iris')
df = iris.drop(['species'],axis=1)
df.head()
```

```python
# kde = kernal density estimation
scatter_matrix(df,figsize=(10,8),diagonal='kde',color='b')
plt.show()
```

# Categoricl Data

In the previous video we have plotted graph where we used numerical data like continuous data vs another set of continues data but when it comes to categorical plotting we use a category data vs numerical data.

The new catplot function provides a new framework giving access to several types of plots that show relationship between numerical variable and one or more categorical variables, like boxplot, stripplot and so on. Catplot can handle 8 different plots currently available in Seaborn.

It is a good complement to a boxplot or violinplot in cases where all observations are shown along with some representation of the underlying distribution. It is used to draw a scatter plot based on the category.

A swarm plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

catplot()

boxplot()

stripplot()

swarmplot()


```python
# catplot
sns.catplot(x='day', y = 'total_bill', data = tips)
```

```python
# swarmplot
sns.catplot(x='day', y = 'total_bill', data = tips)
```

```python
# strip plot
sns.stripplot(x='day', y = 'total_bill', data = tips)
```

```python
# strip plot
sns.stripplot(x='smoker', y = 'total_bill', data = tips)
```

# Dist kde Joint Plot

Now we are aware of various plotting techniques like numerical dataplotting, categorical data plotting etc, In the next video we will understand the KDE plot, distribution plot and will discuss how the data distribution will help us to interpret our data.

distplot() function is used to plot the distplot. The distplot represents the univariate distribution of data i.e. data distribution of a variable against the density distribution. The seaborn. distplot() function accepts the data variable as an argument and returns the plot with the density distribution.

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.
